```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('lr00_results.csv')
data1.head()
```

```python
plt.scatter(data1.x, data1.y, marker='x')
plt.show()
```

```python
def line (m, x, c):
    return (m + x) + c

x = np.linspace(0, 100, 201)
y = line(2, x, 32)
plt.axis ([0, 100, 0, 200])
plt.xlabel('x')
plt.ylabel('y = f(x)')
plt.plot(x, y, c= 'r')
plt.scatter(data1.x, data1.y, marker='x')
plt.show()
```
